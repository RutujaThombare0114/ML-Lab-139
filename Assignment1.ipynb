```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, y='Assessed Improved Value (parcel)')
plt.title('Distribution of Assessed Improved Value')
plt.ylabel('Assessed Improved Value')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Year Built (parcel)'], bins=20, kde=True, color='blue')
plt.title('Year Built Distribution')
plt.xlabel('Year Built')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Longitude', y='Latitude', hue='* CAL FIRE Unit')
plt.title('Incident Locations by CAL FIRE Unit')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(title='CAL FIRE Unit')
plt.show()
```

```python
hazard_counts = data['Hazard Type'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(hazard_counts, labels=hazard_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Proportions of Hazard Types')
plt.show()
```
